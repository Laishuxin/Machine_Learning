# Model Representation

$x^{(i)}$: input value 输入值
$y^{(i)}$: output value, or target features  目标值  
$(x^{(i)}, y^{(i)})$: training example  训练函数  
$m$: training set 训练集，样本数量   
$hypothesis$: 假设函数

$$h_{\theta}(x) = \theta_0 + \theta_1 x$$

# Cost Function（代价函数，均方误差函数）


We can measure the accuracy of the hypothesis function by using a **cost function**

$$J(\theta_0, \theta_1) = \dfrac {1}{2m} \displaystyle \sum _{i=1}^m \left ( \hat{y}_{i}- y_{i} \right)^2 = \dfrac {1}{2m} \displaystyle \sum _{i=1}^m \left (h_\theta (x_{i}) - y_{i} \right)^2$$

验证假设函数好坏的判断依据是：**假设值于真值值之间的差尽可能地小**(例如，我们要预估房出售价，那么我们预估的价格与真实出售的价格要尽可能地接近。即，差值尽可能地小)。  
不仅要预测一个值，而且要预测所有的值。(如，预测不同尺寸的房子的售价)但是，如果简单粗暴地把所有的偏差加起来是不行的。因为，偏差有正有负，正负抵消会让模型失真。  
所以，对差值进行平方。即($h_\theta (x_{i}) - y_{i}^2$)。  
而且这里是取m个样本，所以前面也要乘上1/m


之所以前面乘1/2，是为了后面方便计算梯度。
>The mean is halved $(\frac{1}{2})$ as a convenience for the computation of the gradient descent, as the derivative term of the square function will cancel out the $\frac{1}{2}$term.

假设函数的意义

通过假设函数，我们可以画出代价函数(抛物线)

我们要做的就是找到抛物线的最低点，因为在最低点处，假设函数离散点的均方差和最小

>as a goal, we should try to **minimize the cost function**. In this case, $\theta_1 = 1$ is our global minimum.

### 等高线图(contour plot)

>等高线图含有很多等高线。每一条等高线由两个变量组成的函数，等高线上的值都相等。
