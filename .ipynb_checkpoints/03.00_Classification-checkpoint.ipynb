# Classification


>进行分类的方法有很多，其中一种就是通过线性回归的方式进行预测。通过映射，如果一个值大于0.5我们可以把它分为a_1类，如果一个值小于0.5我们可以把它分为a_2类。但实际上这种方法不是很好，因为分类事实上被不是简单的线性关系。

## Binary classification problem

>分类问题其实与回归问题类似，这是我们的预测结果是一个或者几个离散变量。

>二元分类：只有两个输出结果(0/1)。  

例如：我们对垃圾邮件进行分类。把正常邮件的值设置为1, 把垃圾邮件的值设置为0。

因为输出的y∈{0，1}。0也可以称做**负类**，1称做**正类**。或者我们可以用+、-表示

# Hypothesis Regresentation

我们可以采用假设回归的方式进行预测分类。但这种方法构造得到的结果很差，很容易受到偏离值的影响。不过我们可以修改我们的假设函数，让它的结果输出在[0, 1].

我们修改我们的假设函数的输入值$\theta^Tx$的映射关系，从而实现输出值落在[0, 1]。($0≤h_{\theta}(x)≤1$)

## Logistic Function(Sigmoid Function、逻辑回归函数)

$\begin{align*}& h_\theta (x) = g ( \theta^T x ) \newline \newline& z = \theta^T x \newline& g(z) = \dfrac{1}{1 + e^{-z}}\end{align*}$

$g(\theta^Tx) = \dfrac{1}{1+e^{-\theta^Tx}}$

下面是逻辑回归函数的图像：

通过修改假设函数，使得假设函数区间落在[0，1]。  
$h_\theta(x)$输出结果的意义是xxx被认为是分类1的可能性。


# Decision Boundary(决策边界)

为了把离散的数据设置为[0 ,1]分类，我们可以对假设函数进行转换，我们把假设函数输出值大于等于0.5的归为分类1，把小于0.5的归为分类0.

\begin{align*}& h_\theta(x) \geq 0.5 \rightarrow y = 1 \newline& h_\theta(x) < 0.5 \rightarrow y = 0 \newline\end{align*}

我们可以将上面的式子回代到之前所给我logistic function，可以得到如下：

\begin{align*}& g(\theta^Tx) \geq 0.5 \newline&when \;z \geq 0 \end{align*} 

这里的公式可以进行如下推导:

\begin{align*}& g(\theta^Tx) = \dfrac{1}{1 + e^{-\theta^Tx}} \geq 0.5 \end{align*}
$$e^{-\theta^Tx} \leq 1 \Leftrightarrow z = \theta^Tx \geq0$$

$\Rightarrow$  
$\theta^Tx \geq 0 \Rightarrow y = 1$  
$\theta^Tx \lt 0 \Rightarrow y = 0$  



>决策边界:就是那条把输入值分为y=0部分和y=1部分的线

# Cost Function

如果沿用之前的代价函数，则得到的函数不是凸函数，而且是波浪线的，这样得到的代价函数值会有很多局部最小值。因为我们需要修改代价函数，让它符合逻辑又能收敛。


下面是更新后的代价函数：

\begin{align*}& J(\theta) = \dfrac{1}{m} \sum_{i=1}^m \mathrm{Cost}(h_\theta(x^{(i)}),y^{(i)}) \newline & \mathrm{Cost}(h_\theta(x),y) = -\log(h_\theta(x)) \; & \text{if y = 1} \newline & \mathrm{Cost}(h_\theta(x),y) = -\log(1-h_\theta(x)) \; & \text{if y = 0}\end{align*}

这样修改后的函数：  
对于y=1时，   
$J(\theta)=0 \; when \; y=1$：即我们预测对了，相应的偏差就是最小。  
$J(\theta) = \infty \; when \; y=0$：即我们预测错了，相应的偏差就会特别大。这提醒我们还需要进行梯度下降。


对于y=0时，  
$J(\theta) = \infty \; when \; y=1$：即我们预测错了了，相应的偏差会特别大。  
$J(\theta) = 0 \; when \; y=0$： 即我们预测对了，相应的偏差就为0。

# Simplified Cost Function and Gradient Descent

\begin{align*}& J(\theta) = \dfrac{1}{m} \sum_{i=1}^m \mathrm{Cost}(h_\theta(x^{(i)}),y^{(i)}) \newline & \mathrm{Cost}(h_\theta(x),y) = -\log(h_\theta(x)) \; & \text{if y = 1} \newline & \mathrm{Cost}(h_\theta(x),y) = -\log(1-h_\theta(x)) \; & \text{if y = 0}\end{align*}

我们可以把上面代价函数合成一个函数：  
$$\mathrm{Cost}(h_\theta(x),y) = - y \; \log(h_\theta(x)) - (1 - y) \log(1 - h_\theta(x))$$

可以看到。当我们把y=0，y=1代进去就得到对应的代价函数。

所以，完整的J函数如下：

用向量表示如下：
\begin{align*} & h = g(X\theta)\newline & J(\theta) = \frac{1}{m} \cdot \left(-y^{T}\log(h)-(1-y)^{T}\log(1-h)\right) \end{align*}

## Gradient Descent梯度下降

\begin{align*}& Repeat \; \lbrace \newline & \; \theta_j := \theta_j - \dfrac{\alpha}{m} \dfrac{\partial}{\partial \theta_j}J(\theta) \newline & \rbrace\end{align*}

对J求偏导得到如下公式：

\begin{align*} & Repeat \; \lbrace \newline & \; \theta_j := \theta_j - \frac{\alpha}{m} \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)} \newline & \rbrace \end{align*}

## 解决过拟合

1. 减少特征变量。
    + 手动建设特征变量
    + 使用模型选择算法
2. 正则化
    + 保留所有的特征，但控制参数$\theta_j$的大小
    + 当我们有许多稍微有用的特征变量时，正则化的效果会很好。
