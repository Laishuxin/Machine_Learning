# Multiple Features


## 约定的符号

$x^{(i)}_j = $ 训练数据第i行第j个元素  
$x^{(i)} = $  训练数据第i行的  
m = 训练数据的个数  
n = 变量的个数(如，房子的尺寸，离市中心的远近等)


## Hypothesis Funcion

$$h_\theta (x) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \theta_3 x_3 + \cdots + \theta_n x_n$$

我们设置$x_0 = 1$，可以得到

$$\begin{align*}h_\theta(x) =\begin{bmatrix}\theta_0 \hspace{2em} \theta_1 \hspace{2em} ... \hspace{2em} \theta_n\end{bmatrix}\begin{bmatrix}x_0 \newline x_1 \newline \vdots \newline x_n\end{bmatrix}= \theta^T x\end{align*}$$

# Gradient Descent For Multiple Variables

由代价函数我们可以推出：    

$\begin{align*} & \text{repeat until convergence:} \; \lbrace \newline \; & \theta_0 := \theta_0 - \alpha \frac{1}{m} \sum\limits_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) \cdot x_0^{(i)}\newline \; & \theta_1 := \theta_1 - \alpha \frac{1}{m} \sum\limits_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) \cdot x_1^{(i)} \newline \; & \theta_2 := \theta_2 - \alpha \frac{1}{m} \sum\limits_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) \cdot x_2^{(i)} \newline & \cdots \newline \rbrace \end{align*}$

相当于

$\begin{align*}& \text{repeat until convergence:} \; \lbrace \newline \; & \theta_j := \theta_j - \alpha \frac{1}{m} \sum\limits_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) \cdot x_j^{(i)} \; & \text{for j := 0...n}\newline \rbrace\end{align*}$

# Gradient Descent in Practice I - Feature Scalling(特征缩放)

## 为什么要进行特征缩放
1. 如果数据的波动范围很大，那么利用梯度下降的速度会很慢。  
2. 如果变量分布不均匀的话，梯度下降也很难收敛到最优解

>特征缩放可以调整数据，使数据更加集中。理想的数据范围在-1到1之间。事实上，只要数据波动不要太大，梯度下降的效果还是很明显。

>The way to prevent this is to modify the ranges of our input variables so that they are all roughly the same. Ideally:

## 什么是特征缩放

>特征缩放就是把使用输入值除以它们的极差得到新的数据。  
Feature scaling involves dividing the input values by the range (i.e. the maximum value minus the minimum value) of the input variable, resulting in a new range of just 1. 

第二种调整数据的方法是**均值归一化(或者叫做标准化)**，其操作如下：

$x_i := \frac{x_i - \mu_i}{s_i}$  
其中，$\mu_i, s_i$分别是样本(输入数据)的平均值和方差，这个和数理统计一样

# Gradient Descent in Practice II - Learning Rate

**NOTE:**学习速率非常重要，数学上已经证明，只要learning rate足够小，梯度下降就会收敛到某一值。但是，如果learning rate太小，其收敛的速度会很慢，效率不是很高。如果learning rate太大的话，代价函数可能不会收敛，反而是发散的。

## 选取learning rate的几种调试方式

### Plot the cost function

绘制代价函数的曲线图。 x轴为迭代次数，y轴为代价函数值。如果图像是**上升**的，说明选错了learning rate，这时候需要**减小**learning rate。

### Automatic convergence test

设置阈值。可以计算两次收敛的代价函数之差，如果小于某个值则代价函数已经收敛到最小。不过这在实际操作中存在困难。  
>Declare convergence if $J(θ)$ decreases by less than E in one iteration, where E is some small value such as $10^{−3}$. However in practice it's difficult to choose this threshold value.

## Normal Equation(正态方程)

>Gradient descent gives one way of minimizing J. Let’s discuss a second way of doing so, this time performing the minimization explicitly and without resorting to an iterative algorithm. In the "Normal Equation" method, we will minimize J by explicitly taking its derivatives with respect to the θj ’s, and setting them to zero. This allows us to find the optimum theta without iteration. The normal equation formula is given below: 

$\theta = (X^T X)^{-1}X^Ty$

特征方程的要求是：$X^TX$可以转置的。   
如果$X^TX$不可以转置的可能原因：  
+ 多余的特征。比如，两个特征的值非常接近，非常相关，相当于只能算是一个特征。这时可以删除其中一个特征。
+ 特征太多(如，m≤n)。这时需要删除一些特征，或者使用正则化


正态方程不需要进行特征缩放

正态方程和梯度下降都有各自的优缺点。  如何取舍取决于特征变量的个数(一般达到10,000采取迭代方法会好一些)。
