## Grandient Descent Algorithm

The gradient descent algorithm is:  
**repeat until convergence:**（不断**同步更新**，直到收敛）  
$\theta_j := \theta_j - \alpha \frac{\partial}{\partial \theta_j} J(\theta_0, \theta_1)$  
j=0,1 represents the feature index number.
  

梯度下降的核心的：保持$\theta_0, \theta_1$同步更新。

At each iteration j, one should **simultaneously** update the parameters$\theta_1, \theta_2,...,\theta_n$ Updating a specific parameter prior to calculating another one on the $j^{(th)}$ iteration would yield to a wrong implementation. 

(temp0 := xxx 赋值语句，将xxx赋值到temp0上)

## More about info of parameters

1. $\alpha$(learning rate)：可以把$\alpha$想象成下山的步子，通过调整不同步子的大小，可以调整我们下山的速度。  
>**NOTE:**如果$\alpha$太大，相当于我们下山的速度过快，这样可能导致我们到达山 底后还继续前行。  
如果$\alpha$太小，相当于我们下山的速度太慢，这样浪费的时间会很多。  
其两种情况的表现可以通过下图看出。

2. $\frac{\partial}{\partial \theta_j}J(\theta_0, \theta_1)$：可以把这个参数想象成我们下山的方向。只要方向对了，到达山底的可能性就更高

## How does gradient descent converge with a fixed step size $\alpha$?(梯度下降如何调整收敛的速度)

>The intuition behind the convergence is that $\frac{d}{d\theta_1} J(\theta_1)$ approaches 0 as we approach the bottom of our convex function. At the minimum, the derivative will always be 0 and thus we get:

$$\theta_1 := \theta_1 - \alpha * 0$$

简单地说：当代价函数(cost function)达到最小时，左右两边都是渐变小的。也就是说，其偏导数的绝对值会在其等于0处慢慢变小。  
在$\alpha * \frac{\partial}{\partial \theta_j}J(\theta_0, \theta_1)$中，$\alpha$保持不变，而$\frac{\partial}{\partial \theta_j}J(\theta_0, \theta_1)$不断变小，这样就可以调整收料的速度。  
至于方向的控制，可以通过导数符号(正负)来调整

## Substitute our actual cost function and actual hypothesis function

$$h_{\theta}(x_i) = \theta_1 x_i + \theta_0$$

$$J(\theta_0, \theta_1) = \dfrac {1}{2m} \displaystyle \sum _{i=1}^m \left ( \hat{y}_{i}- y_{i} \right)^2 = \dfrac {1}{2m} \displaystyle \sum _{i=1}^m \left (h_\theta (x_{i}) - y_{i} \right)^2$$

对其求偏导数就可以得到如下式子：

$\begin{align*} \text{repeat until convergence: } \lbrace & \newline \theta_0 := & \theta_0 - \alpha \frac{1}{m} \sum\limits_{i=1}^{m}(h_\theta(x_{i}) - y_{i}) \newline \theta_1 := & \theta_1 - \alpha \frac{1}{m} \sum\limits_{i=1}^{m}\left((h_\theta(x_{i}) - y_{i}) x_{i}\right) \newline \rbrace& \end{align*}$
