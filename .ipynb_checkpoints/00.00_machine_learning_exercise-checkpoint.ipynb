```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
```

```python
data = pd.read_csv('machine-learning-ex/ex1/ex1data1.txt', header=None,
            names=['Population', 'Profit'])
ax = data.plot.scatter(x='Population', y='Profit', c='r', marker='x')
```

```python
fig = plt.figure()
ax = plt.axes()
plt.plot(data['Population'], data['Profit'], 'rx', markersize=10)
ax.set_xlim(4, 24)
plt.xticks(np.arange(4, 26, 2))
plt.yticks(np.arange(-5, 30, 5))
plt.show()
```
