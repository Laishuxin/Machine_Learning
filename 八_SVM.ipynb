## 解决样本不平衡问题：重要参数class_weight

想象一个极端的情况：如果模型分类明显不平衡，比如多数类占了98%，少数类只占了2%，那么，无论模型怎么训练，训练结果都会往多数类靠。   
如果我们想要捕捉少数类的话，这样的模型就没有什么意义。


要解决样本不平衡问题，我们可以通过上采样/下采样(一般不采取下采样 )等方式实现，但是对于svm这种计算量特别大的算法而言，上采样的计算成本太大了。  
不过我们可以在sklearn中调整class_weight来解决样本不平衡的问题。sklearn中class_weight的本质是通过输入的不同权重，对不同标签的样本实行不同的**C惩罚项**。

在sklearn中，class_weight可以通过字典{'标签值1'：weight, '标签值2': weight}或者'balance'实现。 

当我们输入{'标签值1'：weight, '标签值2': weight}, sklearn就会帮助我们修改不同标签的C值，如，  
标签值1的C = weight * C，其中C是我们传入模型的参数，默认为1。  

如果我们输入的是'balance'的话，这个模式使用y值自动调整与输入数据的类频率成反比的权重为$\dfrac{n\_samples}{n\_classes * n\_bincount(y)}$

svc的接口fit的参数：sample_weight
每个样本在fit时的权重，让权重 * 每个样本对应的C值来迫使分类器强调设定的权重更大的样本。通常，较大的权
重加在少数类的样本上，以迫使模型向着少数类的方向建模
通常来说，这两个参数我们只选取一个来设置。如果我们同时设置了两个参数，则C会同时受到两个参数的影响，
即 class_weight中设定的权重 * sample_weight中设定的权重 * C。



```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
```

```python
plt.rcParams['figure.dpi'] = 150
```

```python
class_1 = 500
class_2 = 50
centers = [[0, 0], [2, 2]]
cluster_std = [1.5, 0.5]

X, y = make_blobs(n_samples=[class_1, class_2], centers=centers, cluster_std=cluster_std, random_state=0, shuffle=False)
```

```python
(y==1).sum()
```

```python
np.bincount(y)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50, c=y, cmap='rainbow')
```

##### 在数据集上建模

```python
clf = SVC(kernel='linear', C=1.0)
clf.fit(X, y)

wclf = SVC(kernel='linear', class_weight={1:10, 0:1})
wclf.fit(X, y)

display(clf.score(X, y), wclf.score(X, y))

```

可以看到，不带权重的svc模型要比带了权重的svc模型的accuracy分数要高

##### 绘制决策边界

```python
def plot_decision_bound(model_1, model_2, X, y):
    fig = plt.figure()
    plt.scatter(X[:, 0], X[:, 1], s=50, c=y, cmap='rainbow')
    ax = plt.gca()
    xlim = ax.get_xlim()   # xlim = (min, max)
    ylim = ax.get_ylim()    
    x_mesh, y_mesh = np.mgrid[xlim[0]:xlim[1]:200j, ylim[0]:ylim[1]:200j]  # 相当于np.linspace(xlim_min, xlim_max, 200) 再进行meshgrid
    xy = np.c_[x_mesh.ravel(), y_mesh.ravel()]
    
    # 计算出决策边界
    bound = model_1.decision_function(xy).reshape(x_mesh.shape)
    wbound = model_2.decision_function(xy).reshape(x_mesh.shape)
    
    a = plt.contour(x_mesh, y_mesh, bound, linestyles=['-'], colors=['k'], levels=[0], alpha=0.7)
    b = plt.contour(x_mesh, y_mesh, wbound, linestyles=['-'], colors=['red'], levels=[0], alpha=0.7)    
    plt.legend([a.collections[0], b.collections[0]], ['non-weighted', 'weighted'], loc='best')
    plt.show()
    
```

```python
plot_decision_bound(clf, wclf, X, y)
```

很明显可以看到，没有带权重的模型对少数类不敏感，很多少数类并没有分对。  
而带了权重的模型能很好的捕抓到少数类，基本上所有的少数类都被划分正确，但同时也误伤了很多正类

```python
print(clf.score(X, y))
print(wclf.score(X, y))
```

如果我们先捕抓更多的少数类，我们就难免误伤多数类，由于样本不均衡，带权重的模型评分势必比没带权重的模型评分来得低

如果我们一心想捕抓少数类，那我们宁可牺牲一点模型的精确度。

# 模型的评估指标

## 混淆矩阵

混淆矩阵是二分类问题的多维衡量指标体系，在样本不平衡时极其有用。在混淆矩阵中，我们将少数类认为是正
例，多数类认为是负例。在决策树，随机森林这些普通的分类算法里，即是说少数类是1，多数类是0。在SVM里，
就是说少数类是1，多数类是-1。普通的混淆矩阵，一般使用{0,1}来表示。

![](images/8_2.png)

其中，1，0代表着标签值。**少数类为1， 多数类为0**

在混淆矩阵中，我们明显可以看出，11，00对角线上的样本就是我们预测正确的。01，10对角线上的样本都是我们预测错误的。  
所以，所有以11，00为分子的指标越接近1越好。  
所有以01，10为分子的指标越接近于0越好

混淆矩阵的各种指标：  
1. 模型整体效果：准确度
$$ACCURACY = \dfrac{11 + 00}{11 + 10+ 01+  00}$$

2. 捕抓少数类
    + 精确度:正类样本中预测正确的占所有正类样本的比例。精确度越高，说明少数类预测得越准。
    $$PRECISION = \dfrac{11}{11 + 01}$$
    + 召回率(recall)：对于少数类，预测正确的占预测为少数类的比例。召回率越高，说明我们捕捉了更多的少数类。

    $$RECALL = \dfrac{11}{11 + 10}$$
    + F1-measure：为了同时兼顾精确度和召回率，我们创造了两者的调和平均数作为考量两者平衡的综合性指标。两个数之间的调和平均倾向于靠近两个数中比较小的那一个数，因此我们追求尽量高的F1 measure，能够保证我们的精确度和召回率都比较高。F1 measure在[0,1]之间分布，越接近1越好。  
    
    $$F1-measure = \dfrac{2}{\dfrac{1}{PRECISION} + \dfrac{1}{RECALL}} = \dfrac{2 \times PRECISION * RECALL}{PRECISION + RECALL}$$ 

```python
# 计算模型的精确度
print((y[clf.predict(X) == 1] == 1).sum() / (clf.predict(X) == 1).sum())
print((y[wclf.predict(X) == 1] == 1).sum() / (wclf.predict(X) == 1).sum())
```

可以看到，没带权重的模型的精确度比带了权重的模型的精确度要高。  
这是因为带权重模型错误地把很多多数类预测错了，导致PRECISION的分母变大(虽然分子也变大，但是分子增加的幅度要比分母增加的幅度来得小)


```python
# 计算模型的召回率
print((y[(clf.predict(X) == 1)] == 1).sum() / (y==1).sum())
print((y[(wclf.predict(X) == 1)] == 1).sum() / (y==1).sum())
```

可以看到没带权重的模型的召回率要比带了权重的模型的召回率低得多。

3. 判错多数类的考量：特异度与假正率

+ 特异度(SPECIFICITY)：表示所有真实为0的样本中，被正确预测为0的样本所占的比例
$$SPECIFICITY = \dfrac{00}{00 + 01}$$
+ 假正率（False Positive Rate）:1 - specificity就是一个模型将多数类判断错误的能力
$$FPR = 1 - SPECIFICITY$$

## ROC及其相关

ROC曲线的衡量平衡就是在模型尽量捕抓少数类的时候，观察多数类是如何变化的。  

原理：在我们追求更高的召回率(recall)时，模型的精确度（precision）势必会下降，也就是说，我们捕抓了更多少数类的同时，误伤了多数类。但是如果我们想要知道随机recall的增加，模型将多数类判断错误的程度由多大。假正率(FPR)就可以帮我们得出结论。我们可以使用recall与FPR之间的平衡代替我们对recall与precision之间的平衡。

有些算法天生就带有概率，比如逻辑回归与朴素贝叶斯。但是有些算法如决策树，SVM本身就不带概率，但是可以采取其他指标来计算概率。  
在决策树种，对于每一个叶子节点可能有不同的分类，我们可以用某一种类占所有叶子节点样本个数的比例作为概率，但是某些叶子节点的不纯度为0，即用概率预测的话要么是0要么是1，我们就无法用阈值调整模型。

### SVM实现概率预测：重要参数probability，接口predict_proba以及decision_function

在SVM中，概率是通过样本点到超平面的距离算出来的。  接口decision_function也可以认为是SVM的置信度(confidence)

SVM中probability默认为Fasle，改为True就可以使用接口预测概率

```python
#使用最初的X和y，样本不均衡的这个模型
class_1 = 500 #类别1有500个样本
class_2 = 50 #类别2只有50个
centers = [[0.0, 0.0], [2.0, 2.0]] #设定两个类别的中心
clusters_std = [1.5, 0.5] #设定两个类别的方差，通常来说，样本量比较大的类别会更加松散
X, y = make_blobs(n_samples=[class_1, class_2],
        centers=centers,
        cluster_std=clusters_std,
        random_state=0, shuffle=False)
#看看数据集长什么样
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="rainbow",s=10)
#其中红色点是少数类，紫色点是多数类
```

```python
clf_proba = SVC(kernel="linear",C=1.0,probability=True).fit(X,y)
```

```python
clf_proba.predict_proba(X)
```

```python
clf_proba.decision_function(X)
```

毋庸置疑，Platt缩放中涉及的交叉验证对于大型数据集来说非常昂贵，计算会非常缓慢。另外，由于Platt缩放的理
论原因，在二分类过程中，有可能出现predict_proba返回的概率小于0.5，但样本依旧被标记为正类的情况出现，
毕竟支持向量机本身并不依赖于概率来完成自己的分类。如果我们的确需要置信度分数，但不一定非要是概率形式
的话，那建议可以将probability设置为False，使用decision_function这个接口而不是predict_proba

### 绘制SVM的ROC曲线

```python
# 计算模型的recall与FPR
cm = CM(prob['y_true'], prob['pred'], labels=[1, 0])
recall = R(prob['y_true'], prob['pred'])
recall
```

```python
# FPR
fpr = cm[1, 0] / cm[1, :].sum()
fpr
```

```python
# 通过调节不同的阈值绘制svm不同的学习曲线

prob_range = np.linspace(clf_proba.predict_proba(X).min(), 
                         clf_proba.predict_proba(X).max(), num=50, endpoint=False)
```

```python
recall_list = []
fpr_list = []
for prob in prob_range:
    y_predict = (clf_proba.predict_proba(X)[:, 1] > prob).astype(np.int8)
    cm = CM(y, y_predict, labels=[1, 0])
    recall_list.append(R(y, y_predict))
    fpr_list.append(cm[1, 0] / cm[1:].sum())
```

```python
recall_list.sort()
fpr_list.sort()
```

```python
plt.plot(fpr_list, recall_list, c='red')
plt.plot(prob_range, prob_range, linestyle='--', c='black')
plt.ylabel('recall')
plt.xlabel('FPR')
plt.show()
```

对上图的理解：横坐标为FPR，代表着误伤多数类的能力，纵坐标为recall，代表着预测少数类的能力。中间的虚线是我们捕抓一个少数类的同时误差一个多数类。  

我们绘制ROC曲线的目的是想找出一个阈值，在捕抓更多少数类的同时，尽可能的误伤更少的多数类(即，recall越大且FPR越小)。所有ROC曲线越偏向左边越好。  
如果ROC曲线靠近右边的话，要么是我们预测错误，要么是标签打错了。 但是ROC曲线靠近中间的虚线，那么说明模型的效果很差。  

为了让recall越大，FPR越小，可以引入一个指标AUC(area under curve)，ROC曲线下方的面积。  
AUC越大，模型的效果越好


### sklearn中ROC曲线与AUC面积

```python
from sklearn.metrics import roc_curve 
```

```python
FPR, recall, threshold = roc_curve(y, clf_proba.decision_function(X), pos_label=1)
```

```python
plt.plot(FPR, recall)
plt.plot(np.linspace(0, 1, 50),np.linspace(0, 1, 50), c='red', linestyle='--')
```

```python
from sklearn.metrics import roc_auc_score

area = roc_auc_score(y, clf_proba.decision_function(X))
```

```python
area
```

```python
plt.figure()
plt.plot(FPR, recall, color='red',
label='ROC curve (area = %0.2f)' % area)
plt.plot([0, 1], [0, 1], color='black', linestyle='--')
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('Recall')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

### 利用ROC曲线找出最佳阈值

由ROC曲线，我们可以看出最佳阈值对于的点就是在左上角某一个离虚线最远的点。也是recall-FPR最大的点。这个点也称约登指数。

```python
threshold[(recall - FPR).argmax()]
```

```python
# 在图像上表示出来 
max_idx = (recall - FPR).argmax()
plt.scatter(FPR[max_idx], recall[max_idx], s=30, c='black')

area = roc_auc_score(y, clf_proba.decision_function(X))
label = 'auc = {:.2f}'.format(area)
plt.plot(FPR, recall, linestyle='-', c='red', label=label)  # 画出roc曲线
plt.plot(np.linspace(0, 1, 50), np.linspace(0, 1, 50), c='black', linestyle='--')  # 画出中间的虚线，用于比较

plt.xlabel('FPR')
plt.ylabel('recall')
plt.legend(loc='best')
plt.text(FPR[max_idx], recall[max_idx]+0.02, 'threshold={:.4f}'.format(threshold[max_idx]))
plt.show()
```

可以看到日期是有重复值的，其中日期多大2141个分类。
我们不可能按照日期进行分类，因为这样需要产生多大2141个分类，而且日期不是连续值，只能算是离散数据。如果按照日期进行分类需要产生更多的特征，这样的模型的计算量势必会大大增大。   

我们知道今天下雨可能会影响明天是否下雨，直接删除日期有可能影响模型的精确度，不过我们可以创造一个特征，让它来代替我们的日期。一般而言，下雨可能按照月份或者季度呈现规律，我们可以用月份代替当前的日期。  
同时，我们可以看到在特征中，有一个"Rainfall"，今日降雨量，我们可以设定特定的阈值，如果Rainfall超过这个阈值的话，我们就认为今天下雨了。很多时候今天是否下雨对明天会有一定的影响，所以可以按照Rainfall再创造一个新特征来取代日期

计算样本所在的Location距离澳大利亚主要城市的距离

```python
# 先把角度转换为弧度制
citylld.loc[:, 'slat'] =  citylld['Latitudenum'].apply(np.radians)  
citylld.loc[:, 'slon'] =  citylld['Longitudenum'].apply(np.radians)  
```

```python
# 先把角度转换为弧度制
sample_cityd.loc[:, 'elat'] =  sample_cityd['Latitudenum'].apply(np.radians)  
sample_cityd.loc[:, 'elon'] =  sample_cityd['Longitudenum'].apply(np.radians)  
```

```python
R = 6371.01  # 地球半径
for i in range(sample_cityd.shape[0]):
    slat = citylld['slat']
    slon = citylld['slon']
    elat = sample_cityd.loc[i, 'elat']
    elon = sample_cityd.loc[i, 'elon']

    dist = np.arccos(np.sin(slat)*np.sin(elat) + np.cos(slat)*np.cos(elat)*np.cos(slon - elon))
    
    min_index = np.argmin(dist)
    sample_cityd.loc[i, 'closest_city'] = citylld.loc[min_index, 'City']
    sample_cityd.loc[i, 'climate'] = citylld.loc[min_index, 'Climate']
    
```

```python
from my_decorator import decorator
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score, recall_score
from sklearn.model_selection import cross_val_score
```

```python
import time
def svc_train(X_train, y_train, X_test, y_test, **kwargs):
    print('start running')
    start = time.time()
    kernel = kwargs.get('kernel', 'linear')
    degree = kwargs.get('degree', 1)
    C = kwargs.get('C', 1)
    gamma = kwargs.get('gamma', 'auto')
    coef0 = kwargs.get('coef0', 0.0)
    probability = kwargs.get('probability', False)
    cache_size = kwargs.get('cache_size', 5000)
    class_weight = kwargs.get('class_weight', None)
    random_state = kwargs.get('random_state', 0)
    
    clf = SVC(kernel=kernel, degree=degree, C=C, gamma=gamma, coef0=coef0, probability=probability, class_weight=class_weight,
             cache_size=cache_size, random_state=random_state).fit(X_train, y_train)
    result = clf.predict(X_test)
    score = clf.score(X_test, y_test)
    
    recall = recall_score(y_test, result)
    auc = roc_auc_score(y_test, clf.decision_function(X_test))
    print("kernel:{}\n accracy socre:{}\n recall_score:{}\n auc:{}".format(kwargs.get('kernel'), score, recall, auc))
    
    end = time.time()
    print('spend time : {}'.format(end - start))
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernels:
    svc_train(X_train, y_train, X_test, y_test, kernel=kernel, cache=5000, degree=1, gamma='auto')
```
