# 基本运算

## 算数运算
+ `/`：右除，正常除法(对于矩阵则是B/A == B\*inv(A))
+ `\`：左除，(A\B 就是B * $A^{-1}$)
运用到矩阵$Ax = b$求解就是:`x = A\b`

## 点运输
`.*、./`等等：都是按位运算

## 关系运算

+ <
+ > 
+ ~=：不等于

矩阵对比返回的是一个同样维度的**逻辑**矩阵
