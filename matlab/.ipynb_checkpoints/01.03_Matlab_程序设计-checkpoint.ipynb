# if语句的选择结构

## 单分支if语句

```
if cond
    statement
end
```

## 双分支if语句

```
if cond
    statement
else
    statement
end
```

## 多分支if语句

```
if cond1
    statement1
elseif cond2
    statement2
elseif cond3
    statement3
...

else
    statementn
end
```

# switch语句的选择结构

```
switch expression
case result1  % 如果result有多个可能，则用数据表表示，如{1，2，4}  
    statement1
case result2
    statement2
...
otherwise
    statementn
end
```

示例2：
PM2.5是指大气中直径小于或等于2.5微米的可入肺颗粒物，是衡量空气质量的重要指标。假定空气质量等级以PM2.5数值划分为6级。PM2.5数值在[0，35）空气质量为优，[35，75）为良，[75，115）为轻度污染，[115，150）为中度污染，[150，250）为重度污染，大于等于250为严重污染。编写程序，输入PM2.5数值，输出空气质量等级
+ 用六个单分支结构实现
+ 用多分支if语句实现
+ 用switch语句执行

```
function pm25(pm)
if pm>=0 & pm<35
    disp('空气质量为优')
end
if pm >=35  pm<75
    disp('空气质量为良')
end
if pm >=75 & pm <115
    disp('轻度污染')
end
if pm >=115 & pm <150
    disp('中度污染')
end
if pm >=150 & pm<250
    disp('重度污染')
end
if pm >= 250
    disp('严重污染')
end
```

```
function pm25_2(pm)
if pm >= 250
    disp('严重污染')
elseif pm >= 150
    disp('重度污染')
elseif pm >= 115
    disp('中度污染')
elseif pm >= 75
    disp('轻度污染')
elseif pm >= 35
    disp('良')
else
    disp('优')
end
```

# 用for语句实现循环

```
for 循环变量 = 表达式1(初值):表达式2(步长):表达式3(终值)
    循环体语句
end
```

退出循环后，循环变量的值就是最后一个元素

示例1：计算圆周率  
$\pi/4 = 1 - 1/3 + 1/5 - 1/7 + \cdots + (-1)^{n+1}*1/(2n - 1)$

```
function pai = computePi(n)
temp = 0;
for i = 1:n
     temp = temp +  (-1)^(i+1) / (2*i -1);
end
pai = temp * 4;
```

示例2：

```
function pai = computepi3(bins)
% bins：整数。要把区间bins小份
a = 0; %积分下限
b = 1;  %积分上限

n = (b - a) / bins;  % 将区间分成n小份
area = 0;  % 面积，用于接收积分面积
for i = a:n:b
    h = sqrt(1 - i^2);  % 长方形的高度
    area = area + n * h;
end
pai = 4 * area
```

示例2：使用蒙特卡洛方法实现pi的计算

```
function pai = computePi5(n)
%使用蒙特卡洛方法实现
s = 0;
for i = 1:n
    x = rand(1);
    y = rand(1);
    if x ^2 + y ^2 <1
        s = s + 1;
    end
end
pai = s / n * 4;
```

# while语句实现循环结构

示例1：从键盘输入若干个数，当输入0时结束输入，求这些数的平均值和它们之和。

```
function [sum_, mean_] = whileSum()
% 利用while 循环实现求n个数和
sum_ = 0;
count = 0;
x = input('请入一个数：(输入0退出)：');
while x ~= 0
    sum_ = sum_ + x;
    count = count + 1;
    x = input('请入一个数：(输入0退出)：');
end
if count ~= 0 
    mean_ = sum_ / count;
    fprintf('和为：%d, 均值为：%d',sum_,mean_);
end
```
