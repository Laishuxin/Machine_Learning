```
function [c, s]=goldPoint()
a = input('请输入黄金分割点a的复数坐标：');
b = input('请输入黄金分割点b的复数坐标：');
s = abs(b - a);
c = a + 0.618*s;
disp(s)
disp(c);
```

# if语句的选择结构

## 单分支if语句

```
if cond
    statement
end
```

## 双分支if语句

```
if cond
    statement
else
    statement
end
```

## 多分支if语句

```
if cond1
    statement1
elseif cond2
    statement2
elseif cond3
    statement3
...

else
    statementn
end
```

# switch语句的选择结构

```
switch expression
case result1  % 如果result有多个可能，则用数据表表示，如{1，2，4}  
    statement1
case result2
    statement2
...
otherwise
    statementn
end
```

示例1：输入一个英文单词，判断它是否以元音字母开头。
```
c = input('请输入一个单词：', 's')
switch c(1)
case {'a','e','i','o','u','A','E','I','O','U'}
disp([c,'以元音开头'])
otherwise
disp([c,'不是以元音开头'])
end
```

示例2：
PM2.5是指大气中直径小于或等于2.5微米的可入肺颗粒物，是衡量空气质量的重要指标。假定空气质量等级以PM2.5数值划分为6级。PM2.5数值在[0，35）空气质量为优，[35，75）为良，[75，115）为轻度污染，[115，150）为中度污染，[150，250）为重度污染，大于等于250为严重污染。编写程序，输入PM2.5数值，输出空气质量等级
+ 用六个单分支结构实现
+ 用多分支if语句实现
+ 用switch语句执行

```
function pm25(pm)
if pm>=0 & pm<35
    disp('空气质量为优')
end
if pm >=35  pm<75
    disp('空气质量为良')
end
if pm >=75 & pm <115
    disp('轻度污染')
end
if pm >=115 & pm <150
    disp('中度污染')
end
if pm >=150 & pm<250
    disp('重度污染')
end
if pm >= 250
    disp('严重污染')
end
```

```
function pm25_2(pm)
if pm >= 250
    disp('严重污染')
elseif pm >= 150
    disp('重度污染')
elseif pm >= 115
    disp('中度污染')
elseif pm >= 75
    disp('轻度污染')
elseif pm >= 35
    disp('良')
else
    disp('优')
end
```
