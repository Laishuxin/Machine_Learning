## Gradient descent

### Update Equations

The objective of linear regression is to minimize the cost function

$$J\left(\theta \right)=\frac{1}{2m}\sum_{i=1}^m {\left(h_{\theta \;} \left(x^{\left(i\right)} \right)-y^{\left(i\right)} \right)}^2$$

where the hypothesis  is given by the linear model

$$h_{\theta } \left(x\right)=\theta^T x=\theta_0 +\theta_1 x_1$$

Recall that the parameters of your model are the  values. These are the values you will adjust to minimize cost . One way to do this is to use the batch gradient descent algorithm. In batch gradient descent, each iteration performs the update

通过批量梯度下降完成任务。  
**NOTE:**一定要同时调整theta

$$\theta_j :=\theta_j -\alpha \frac{\;1}{m}\sum_{i=1}^m \;\left(h_{\theta \;} \left(x^{\left(i\right)} \right)-y^{\left(i\right)} \right)x_j^{\left(i\right)} \;\;\;\;\;\;\;\left(\textrm{simultaneously}\;\textrm{update}\;\theta_{j\;} \;\textrm{for}\;\textrm{all}\;j\right)$$

### Implementation

设置learning_rate:$\alpha=0.01$

```python
m = len(X)
X = np.matrix([np.ones(m), X]).T
y = np.matrix(y.to_numpy()).T
```

```python
theta = np.zeros((2,1))  # 初始化theta[[theta_0], [theta_1]]
iterations = 1500 # 设置迭代次数
alpha = 0.01  # 设置学习步长
```

```python
def computeCost(x, y, theta):
    """
    计算代价函数
    : x: 矩阵。在代价函数中的x_i
    : y: 矩阵。在代价函数中的y_i
    : theta: 矩阵。在代数函数中的θ_0， θ_1
    : return: J 代价函数值
    """
    J = sum(np.square(x * theta - y)) / 2 / m
    return J
```

```python
print(computeCost(X, y, theta))
print(computeCost(X, y, [[-1],[2]]))
```

### Gradient descent

```python
def gradientDescent(X, y, theta, alpha, num_iters):
    """
    计算梯度下降, 一定一定要让theta同时更新
    : x: 矩阵。在代价函数中的x_i
    : y: 矩阵。在代价函数中的y_i
    : theta: 矩阵。在代数函数中的θ_0， θ_1
    : alpha:learning rate学习速率
    : num_iters: 迭代次数
    : return: theta(经过梯度下降后的theta), J_history(J的历史数据) 
    """
    m = len(X)
    J_history = np.zeros((num_iters, 1))
    for i in range(num_iters):
        J_history[i, 0] = computeCost(X, y, theta)
        temp0 = theta[0,0] - alpha * sum(X*theta - y) / m
        temp1 = theta[1,0] - alpha * ((X * theta - y).T * X[:, 1]) / m
        theta[0,0] = temp0
        theta[1,0] = temp1
    return theta, J_history
```

```python
temp0 = theta[0,0] - alpha * sum(X*theta - y) / m
temp1 = theta[1,0] - alpha * ((X * theta - y).T * X[:, 1]) / m
print(temp0, temp1)
```

```python
theta, J_history = gradientDescent(X, y, theta, alpha, iterations)
```

```python
print('根据梯度下降得到的theta为：\ntheta_0:{}\ntheta_1:{}'
      .format(theta[0,0], theta[1,0]))
```

```python
fig = plotData(X[:, 1], y)
plt.plot(X[:, 1], X*theta)
plt.legend(['Training Data', 'Linear regression'], frameon=False)
plt.show()
```
