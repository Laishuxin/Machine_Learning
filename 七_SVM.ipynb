# 支持向量机


SVM的功能：  


SVM对于线性与非线性数据集都有很好的效果。

SVM的优点：  
+ 在高维度空间处理数据非常优秀
+ 对于样本数量比特征维度还少的数据集仍然适用
+ 在决策函数中使用训练点的子集（称为支持向量），因此它也具有内存效率。
+ 可以用不同的核训练多样化的数据
  
SVM的缺点：  
+ 没有返回样本的分类概率

## SVM的工作原理

>SVM的工作原理可以简单总结如下：  
对于线性可分二分类问题，找到一个超平面作为决策边界，使得样本点分布在超平面两侧。然后，让距离超平面最近的点离超平面最远，对应那些离超平面最近的点称之为支持向量。

超平面就是比特征维度低一维的空间。比如一个数据集，它的特征维度为3(三维空间)，那么超平面维度为2(三维空间的二维子空间)

```python
fig = plt.figure()
plt.scatter(X[:,0], X[:,1],c=y, cmap='rainbow', s=50)
ax = plt.gca()
```

```python
"""
1. 获取将要绘制图片的坐标上下界
2. 将获取到的坐标边界分为n个点
3. 将分成的点形成一个坐标网络
4. 画出等高线
"""

xlim = ax.get_xlim()
ylim = ax.get_ylim()
display(xlim, ylim)
```

```python
x_ = np.linspace(xlim[0], xlim[1], 30)
y_ = np.linspace(ylim[0], ylim[1], 30)
```

```python
# 返回样本点到决策边界的距离
z = clf.decision_function(xy).reshape(axisx.shape)
```

```python
# 呼出等高线图, 指定等高线维-1，0，1，等高线的styles为xxx
ax.contour(axisx, axisy, z, levels=[-1, 0, 1], colors='k', alpha=0.5, linestyles=['--', '-', '--'])
ax.set_xlim(xlim)
ax.set_ylim(ylim)
fig
```

#### 将整个过程包装成函数

```python
def plot_svc_decision_function(model, ax=None):
    """
    根据传入的SVC模型进行绘图
    
    Parameters
    ----------
    model : class sklearn.svm.SVC
        训练好的SVC模型
    ax : plt.axes 
        画图对象
    """
    # 1. 判断子图是否已存在，如果存在则获取子图，不存在则创建子图
    if ax is None:
        ax = plt.gca()
    
    # 2. 获取坐标轴区间端点
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # 3. 将坐标区间分为多个点
    n = 30  # 分为30个点
    xs = np.linspace(xlim[0], xlim[1], n)
    ys = np.linspace(ylim[0], ylim[1], n)

    # 4. 形成网络，并形式样本坐标
    x_mesh, y_mesh = np.meshgrid(xs, ys)
    xy = np.vstack([x_mesh.ravel(), y_mesh.ravel()]).T
    # 5. 计算样本坐标到决策边界的距离
    z = model.decision_function(xy).reshape(x_mesh.shape)

    # 6. 绘制等高线图
    plt.contour(x_mesh, y_mesh, z, colors='k', levels=[-1, 0, 1], linestyles=['--', '-', '--'])
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)    
    plt.show()
    
```

```python
clf = SVC(kernel='linear').fit(X, y)

plt.scatter(X[:, 0], X[:, 1], cmap='rainbow', s=50, c=y)
plot_svc_decision_function(clf)
```

模型的接口

```python
# 预测分类
clf.predict(X)
```

```python
# 为模型评分
clf.score(X, y)
```

```python
# 找出模型的支持向量
clf.support_vectors_
```

```python
# 返回每个分类中的支持向量的个数
clf.n_support_
```

## 非线性SVM与核函数

### SVC在非线性数据的推广

为了让非线性数据依然可分，我们需要把数据映射到更高维度的空间中，而这种映射是非线性映射。  


从低维度空间到高维度空间的映射非常巧妙，但也存在很多问题。如：  
更高维度意味着计算量成倍增加，  
有时数据还可能会映射到无限维的空间中。  

为了解决如上的问题所使用的数学方式称之为**核技巧**。

核函数能解决的问题：  
+ 因为非线性SVM中的核函数都是正定核函数(positive definite kernel functions)，他们都满足美世定律(Mercer's theorem)，确保了高维空间中任意两个向量的点积一定可以被低维空间中的这两个向量的某种计算来表示。  
+ 使用核函数计算低维度空间的向量关系比计算高维度空间的计算成本要低很多 
+ 计算在原始空间中进行，所以避免了映射到无限维度的空间


sklearn中的kernel参数：  

|  类型   |      含义      | 解决的问题 |           核函数表达式           | 参数gamma | 参数degree |参数coef0|
| :-----: | :------------: | :--------: | :------------------------------: | :-------: | :--------: | :--------: |
| linear  |   线性核函数   |    线性    |         $K(x, y) = x^Ty$         | No | No | No |
|  poly   |  多项式核函数  |   偏线性   |  $K(x, y) =(\gamma(x·y) + r)^d$  | yes | yes |yes|
| sigmoid | 双曲正切核函数 |   非线性   | $K(x, y) =tanh(\gamma(x·y) + r)$ | yes | No |yes|
|   rbf   |   高斯径向基   |  偏非线性  | $K(x, y) =e^{-\gamma||x- y||^2}$ | yes | No |No|



(其中最常用也是最好用的就是**rbf**， 需要标准化数据集。  
多项式核函数的degree默认为3，即三项式拟合。当degree为1是，多项式拟合就变成了线性拟合)

```python
clf = SVC(kernel='rbf').fit(X, y)
plt.scatter(X[:, 0], X[:, 1], s=10, c=y, cmap='rainbow')
plot_svc_decision_function(clf)
```

### 探索核函数在不同数据集上的表现

##### 导库

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.svm import SVC
from sklearn.datasets import make_blobs, make_circles, make_classification, make_moons
```

```python
n_samples = 100

datasets = [
    make_blobs(n_samples, n_features=2, centers=2, cluster_std=2.5, random_state=1),
    make_circles(n_samples, noise=.2, factor=.5, random_state=1),
    make_classification(n_samples, n_features=2, n_redundant=0, random_state=1),
    make_moons(n_samples, noise=0.2, random_state=1)
]

for X, y in datasets:
    fig = plt.figure(figsize =(5, 4))
    plt.scatter(X[:, 0], X[:, 1], s=50, c=y, cmap='rainbow')
```

##### 构建子图

```python
# 定义行列子图个数
kernels = ['linear', 'poly', 'sigmoid', 'rbf']
nrows = len(kernels)
ncols = len(kernels) + 1

# 创建多子图
fig, axes = plt.subplots(nrows, ncols, figsize=(30, 20))

# 对不同类型数据进行循环
for row, (X, y) in enumerate(datasets):
    # 按行进行绘图
    ax = axes[row, 0]
    # 在第一列的子图上添加标题
    if row == 0:
        ax.set_title('input data')
    # 开始绘图
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=plt.cm.Paired, edgecolors='k', zorder=10)
    ax.set_xticks([])
    ax.set_yticks([])
    # 对其他列进行绘图
    for col, kernel in enumerate(kernels, start=1):
        # 获取子图位置
        ax2 = axes[row, col]
        if row == 0:
            ax2.set_title(kernel)
        # 建模
        clf = SVC(kernel=kernel, gamma=2).fit(X, y)
        # 评分
        score = clf.score(X, y)
        # 对数据本身绘制散点图
        ax2.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=plt.cm.Paired, edgecolors='k', zorder=10)
#         ax2.set_xticks([])
#         ax2.set_yticks([])

        # 画出支持向量
        ax2.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=60, facecolors='none', zorder=10, edgecolors='blue')
        
        # 绘制决策边界
        x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
        y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

        n = 200  # 将坐标轴划分200个点
        xs = np.linspace(x_min, x_max, n)
        ys = np.linspace(y_min, y_max, n)
        
        x_mesh, y_mesh = np.meshgrid(xs, ys)
        
        xy = np.vstack([x_mesh.ravel(), y_mesh.ravel()]).T
        
        z = clf.decision_function(xy).reshape(x_mesh.shape)
        
        ax2.pcolormesh(x_mesh, y_mesh, z>0, cmap=plt.cm.Paired)
        
        ax2.contour(x_mesh, y_mesh, z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'], levels=[-1, 0, 1])
        
        # 添加分数文字
        ax2.text(x_max -0.5, y_min + 0.5, ('%.2f' %score).lstrip('0'), size=15, 
                bbox=dict(boxstyle='round', alpha=.8, facecolor='white'),
                horizontalalignment='right')
        
plt.tight_layout()
plt.show()
```
